```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
fig.suptitle('Distribuição dos atributos', fontsize=20)
sns.histplot(amazon, x='Open', ax=axes[0, 0])
sns.histplot(amazon, x='High', ax=axes[0, 1])
sns.histplot(amazon, x='Low', ax=axes[0, 2])
sns.histplot(amazon, x='Close', ax=axes[1, 0])
sns.histplot(amazon, x='Adj Close', ax=axes[1, 1])
sns.histplot(amazon, x='Volume', ax=axes[1, 2])
plt.show()
```

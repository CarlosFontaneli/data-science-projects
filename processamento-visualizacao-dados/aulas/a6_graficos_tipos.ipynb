```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
t = arange(0.0, 2.0, 0.01)
s = sin(2.5*pi*t)
plot(t, s)
xlabel('time (s)')
ylabel('voltage (mV)')
title('Sine Wave')
grid(True)
show()
```

```python
x=np.arange(10)

plt.plot(x,x*1,c='b',marker="^",ls='--',label='Linear',fillstyle='none')
plt.plot(x, x**2,c='g',marker=(8,2,0),ls='--',label='*2')
plt.plot(x,x**3,c='k',ls='-',label='*3')
plt.title("Taxa de Crescimento de tempo dos algoritmos.")
plt.xlabel("n")
plt.ylabel("Tempo de execucao (segundos)")
plt.legend(loc=2)
plt.show()
```

```python
x=np.linspace(0, 12, 10000)

plt.plot(x,x*1,c='b',ls='-',label='Linear',fillstyle='none')
plt.plot(x, x**2,c='g',ls='-',label='Quadr√°tica')
plt.plot(x,x**3,c='k',ls='--',label='C√∫bica')
plt.plot(x,2**x,c='r',ls='--',label='Exponencial')
plt.title("Taxa de Crescimento de tempo dos algoritmos.")
plt.xlabel("n")
plt.ylabel("Tempo de execucao (segundos)")
plt.legend(loc=2)
plt.show()
```

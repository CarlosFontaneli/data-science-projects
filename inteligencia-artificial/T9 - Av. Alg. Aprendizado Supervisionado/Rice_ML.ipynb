# Importando o DataSet

```python
df = pd.read_csv("Rice_Cammeo_Osmancik.csv",delimiter=";",decimal=",")
df.reset_index(inplace=True,drop=True)
```

# Separar em matrizes de atributos e vetores de classe

```python
X = df.drop('Class',axis=1)
y = df[['Class']]
```

# Separar em 10 folds

A função <b>tenfond_manual</b> irá criar uma lista de índices com 10 elementos, sendo cada elemento um fold

```python
def tenfold_manual(df,seed=420):
    intervalo = np.arange(start = 0, stop = len(df))
    random.Random(seed).shuffle(intervalo) #random.Random(seed) define uma seed para o random usar
    
    return np.array_split(intervalo, 10)
```

A função <b>retorna_fold</b> irá auxiliar os folds a serem acessados durante o for 

```python
def retorna_fold(fold,pos):
    train_index = []
    test_index = folds[pos]
    contador = 0
    for i in folds:
        if contador != pos:
            for j in i:
                train_index.append(j)
        contador += 1
    return train_index,test_index
```

```python
folds = tenfold_manual(df)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib as plt

clf = DecisionTreeClassifier(criterion="entropy")

for n in range(10):
    train, test = retorna_fold(folds,n)
    X_train = X.iloc[train]
    y_train = y.iloc[train]
    
    X_test = X.iloc[test]
    y_test = y.iloc[test]
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)

    print(f"Fold: {n+1} \nMatriz de confusão:\n{confusion_matrix(y_test, y_pred)}\n")
    print(f"Classification Report:\n{classification_report(y_test, y_pred)}\n")
    
```
